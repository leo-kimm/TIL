```python
# 채점 프로그램

a = input('답 : ')
b = input('정답 : ')
aa = list(a)
bb = list(b)
cnt = 0
n = 0
ac = []
num = 1

for ai in aa:
    ac.append(str(num)+'번 *****')
    if ai == bb[n]:
        ac.append('O')
     
           
        cnt+=1
  
        
    else:
        ac.append('--->>   X an:'+bb[n])

    n+=1
    
    
    if num == 20:
        fir=cnt
        print(f'1단원:{cnt}')
    if num == 40:
        sec=cnt
        print(f'2단원:{cnt-fir}')
    if num == 60:
        thd=cnt
        print(f'3단원:{cnt-sec}')
    if num == 80:
        print(f'4단원:{cnt-thd}')   

            
            
    num+=1        
    
print(f'총 점수는 {cnt/80*100}입니다.')
for i in ac:
    print(i)
```
