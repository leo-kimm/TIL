# 오늘의 코딩

```python
# 코딩연습 내적 ver.1

def solution(a, b):
    answer = 0
    for i,j in zip(a,b):
        answer += i*j
    return answer
    
solution([-1,0,1],[1,0,-1])
```

```python
# 코딩연습 내적 ver.2

def solution(a, b):
    answer = 0
    for i,j in zip(range(len(a)),range(len(b))):
        if i == j:
            answer += a[i]*b[j]
    return answer
    
solution([-1,0,1],[1,0,-1])
```

```python
# 두 정수 사이의 합

def solution(a, b):
    answer = 0
    if a<=b:
        for i in range(a,b+1):
            answer += i
    else:
        for i in range(b,a+1):
            answer += i
    return answer

solution(3,5)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
# 날씨 가져오기 1단계 (일단 가장 큰 테이블 명으로 가져오기)
soup.find('table',{'class':'table_develop3'})
```

```python
# tr로 한번 뽑아보기
soup.findAll('tr')

for tr in soup.findAll('tr'):
    print('tr = ', tr)
    print('='*70)
```

```python
# 인덱스 별로 무슨 항목인지, 길이는 몇인지 확인(조건 만들기 위해)
soup.find_all('tr')

for tr in soup.find_all('tr'):
    tds = tr.find_all('td')
    print('tds = \n', tds)
    print('\nlen(tds) = ', len(tds))
    print('*'*70)
    
# 인덱스 별로 동네, 온도, 풍향 등을 의미함
```

```python
# 지역 온도 습도 만 뽑아오기
for tr in soup.find_all('tr'):
    tds = tr.find_all('td')
    if len(tds)>0:
        print('tds[0] : 지역 = ', tds[0].text)
        print('tds[5] : 온도 = ', tds[5].text)
        print('tds[10] : 습도 = ', tds[10].text)
        print('*'*70)
```

```python
# 강의자료 따라하기 p.51
data = []

for tr in soup.find_all('tr'):
    tds = list(tr.find_all('td'))
    for td in tds:
        if td.find('a'):
            point = td.find('a').text
            temper = tds[5].text
            humid = tds[10].text
            data.append([point, temper, humid])
            
data
```

```python
# HTML 가져와서 파일로 저장하기(포맷팅으로 하기)
with open('weather.csv','w') as file:
    file.write('point,temper,humid\n')
    for i in data:
        # {숫자}별로 엑셀 위치 설정 가능
        file.write('{0},{1},{2}\n'.format(i[0],i[1],i[2]))
    print('저장완료!!')
```

```python
# join 쓰면 쉽게 만들수 있음
with open('weather2.csv','w') as file:
    file.write('point,temper,humid\n')
    for item in data:
        row1=','.join(item)
        file.write(row1+'\n')
    print('저장완료!!')
```

```python
# join 연습하기
for item in data:
    print('item : ', item)
    row = ','.join(item)
    print('row : ', row)
    print('*'*70)
```
