```python
# 10952
# A+B - 5


while True:
    a, b = map(int, input().split())
    
    if a == 0 and b == 0:
        break
    else:
        print(a+b)
```

```python
# 1110
# 더하기 사이클

while True:
    try:
        
        n = int(input())
        N = 0

        if n<10:
            n = '0'+str(n)
        n=str(n)
        answer = str(n)



        while True:
            a = n[1]
            b = int(n[0])+int(n[1])
            n = a+str(b)[-1]
            N+=1
            print(f'n={n}')
            print(f'N={N}')
            if n == answer:
                break

        print(N)
    except:
        break
```
