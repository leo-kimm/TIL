```python
# 숫자계산 프로그램 작성하기

nums = list()
while True:
    
    num = input('숫자를 입력하세요 : ')
    
    if num == '입력 끝':
        break
    
    nums.append(int(num))
    
    
# 합 구하기
sum_num = 0    

for i in nums:
    sum_num += i
    
# 평균 구하기
avg_num = sum_num/len(nums)


# 가장 크고 작은 숫자 구하기
nums.sort()
max_num = nums[-1]
min_num = nums[0]


print(f'입력받은 숫자들의 합 : {sum_num}')    
print(f'입력받은 숫자들의 평균 : {avg_num}')
print(f'가장 큰 숫자는 : {max_num}')
print(f'가장 작은 숫자는 : {min_num}')
```

```python
# 숫자계산 프로그램 작성하기

cnt = 0
sum_num = 0
max_num = 0
min_num = 0

while True:
    cnt += 1
    num = input('숫자를 입력하세요 : ')
    
    if num == '입력 끝':
        break
    
    # 문자열 정수형으로 바꿔주기
    i_num = int(num)
    
    # 합 구하기
    sum_num += i_num 
    
    # 가장 큰 숫자 구하기
    if max_num == 0:
        max_num = i_num
    else:
        if max_num < i_num:
            max_num = i_num
    
    # 가장 작은 숫자 구하기
    if min_num == 0:
        min_num = i_num
    else:
        if min_num > i_num:
            min_num = i_num    

    
# 평균 구하기
avg_num = sum_num/cnt


print(f'입력받은 숫자들의 합 : {sum_num}')    
print(f'입력받은 숫자들의 평균 : {avg_num}')
print(f'가장 큰 숫자는 : {max_num}')
print(f'가장 작은 숫자는 : {min_num}')
```

```python
# 실습하기(최댓값 최솟값 찾기)

# 길이가 20인 리스트에 무작위로 1~100 사이의 숫자를 생성하여 저장하자

# 가장 큰 수와 가장 작은 수를 찾아서 출력하시오

# 실제로 가장 큰수와 가장 작은 수를 찾았는지 확인하기 리스트 전체를 확인해보자

# while 사용하기


import random

ran_list = []
i = 0
while i<20:
    r_num = random.randint(1,100)
    ran_list.append(r_num)
    
    i += 1
ran_list.sort()

print(f'가장 큰 수 : {ran_list[-1]}, 가장 작은 수 : {ran_list[0]}')
print(ran_list)
```

```python
# 중첩 for을 이용하여 구구단 2단 부터 9단 만들기

for i in range(2,10):
    print(f'{i}단\n')
    for cnt in range(1,10):
        print(f'{i} x {cnt} = ',i*cnt)
    print()
```

```python
# for 문을 이용하여 숫자자리 중 369가 있으면 그 개수만큼 clap을 출력하는 게임 만들기

num = input('원하는 숫자 주시죠~~~ : \n삼육구~ 삼육구~ \n')
print()
for i in range(1,int(num)+1):
    cnt = 0
    ii = str(i)
    if '3' in str(i) or '6' in str(i) or '9' in str(i):
        for i in range(0,len(ii)):
            if '3' == ii[i] or '6' == ii[i] or '9' == ii[i]:
                cnt += 1
        
        print('짝'*cnt)

    else:
        print(i)
```
