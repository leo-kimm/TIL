```python
# ìˆ«ìê³„ì‚° í”„ë¡œê·¸ë¨ ì‘ì„±í•˜ê¸°

cnt = 0
sum_num = 0
max_num = 0
min_num = 0

while True:
    cnt += 1
    num = input('ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš” : ')
    
    if num == 'ì…ë ¥ ë':
        break
    
    # ë¬¸ìì—´ ì •ìˆ˜í˜•ìœ¼ë¡œ ë°”ê¿”ì£¼ê¸°
    i_num = int(num)
    
    # í•© êµ¬í•˜ê¸°
    sum_num += i_num 
    
    # ê°€ì¥ í° ìˆ«ì êµ¬í•˜ê¸°
    if max_num == 0:
        max_num = i_num
    else:
        if max_num < i_num:
            max_num = i_num
    
    # ê°€ì¥ ì‘ì€ ìˆ«ì êµ¬í•˜ê¸°
    if min_num == 0:
        min_num = i_num
    else:
        if min_num > i_num:
            min_num = i_num    

    
# í‰ê·  êµ¬í•˜ê¸°
avg_num = sum_num/cnt


print(f'ì…ë ¥ë°›ì€ ìˆ«ìë“¤ì˜ í•© : {sum_num}')    
print(f'ì…ë ¥ë°›ì€ ìˆ«ìë“¤ì˜ í‰ê·  : {avg_num}')
print(f'ê°€ì¥ í° ìˆ«ìëŠ” : {max_num}')
print(f'ê°€ì¥ ì‘ì€ ìˆ«ìëŠ” : {min_num}')
```
