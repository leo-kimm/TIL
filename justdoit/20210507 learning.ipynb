```python
import pandas as pd

pd.DataFrame(data=load_iris()['data'], 
            columns = ['sepal length','sepal width','petal length','petal width'])
```

```python
iris_df = pd.DataFrame(data=load_iris()['data'], 
            columns = ['sepal length','sepal width','petal length','petal width'])
```

```python
iris_df['target'] = load_iris()['target']

iris_df
```

```python
X = iris_df.loc[:,'sepal length':'petal width']
```

```python
y = iris_df.loc[:,'target':]
```

```python
from sklearn.model_selection import train_test_split

train_test_split(X,y,test_size=0.2)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,y,test_size=0.2)

print(Y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(oob_score = True)
rfc
```

```python
rfc.fit(X_train, Y_train)
```

```python
rfc.estimators_
```

```python
from  sklearn import tree
import pydotplus
from PIL import Image
from io import BytesIO
```

```python
dot_data = tree.export_graphviz(
    rfc.estimators_[0],
    feature_names = X_train.columns,
    class_names = ['iris setosa','iris versicolor','iris virginica']
)
```

```python
for i in range(1,11):
    dot_data = tree.export_graphviz(
        rfc.estimators_[i],
        feature_names = X_train.columns,
        class_names = ['iris setosa','iris versicolor','iris virginica']
    )
    pydotplus.graph_from_dot_data(dot_data).write_pdf('treeee'+str(i)+'.pdf')
```

```python
predict = rfc.predict(X_test)
predict
```
