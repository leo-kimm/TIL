# 오늘의 코딩

```python
# 평균구하기


def solution(arr):
    
    return sum(arr)/len(arr)

solution([1,2,3,4])
```

```python
# 정수 내림차순으로 배치하기

def solution(n):

    a = sorted(str(n))

    return int(''.join(reversed(a)))

solution(118372)
```

```python
# 문자열 다루기 ver.1

def solution(s):
    if s.isdigit()==True and (len(s)==4 or len(s)==6):
        return True
    else:
        return False
solution("1234")
```

```python
# 문자열 다루기 ver.2

def solution(s):
    try:
        int(s)
    except:
        return False
    return len(s) == 4 or len(s) == 6
solution("1234e")
```

```python
# 수박수박수

def solution(n):
    a=n*'수박'
    return a[:n]

solution(1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# x y 값 설정(~부터, ~까지, ~만큼씩 커짐)
x = np.arange(0, 2, 0.1)
y = np.arange(0, 2, 0.1)
x_1 = np.arange(0, 3, 0.2)
y_1 = np.arange(0, 6, 0.4)
```

```python
#  그래프 만들기 / 그래프 사이즈 설정(가로,세로)
plt.figure(figsize=(5,5))
#plt.plot()
# x,y값 받아오고, 라벨(범례) 내용 설정
plt.plot(x, y, label='y = x')
plt.plot(x_1, y_1, label='y = 2x')
plt.xlabel('x')
plt.ylabel('y')
# x의 표시 범위 제한
plt.xlim(0, 2)
plt.ylim(0, 2)
# x축 표시 기울기 설정
plt.xticks(rotation=45)
plt.yticks(rotation=45)
# 범례 표시 위치 (0=best, 10=center)
plt.legend(loc=0)
# 그리드 표시
plt.grid()
# 그래프 보여주기
plt.show()
```

```python
# 그래프 한번에 여러개 만들때(subplots 사용, 열,행)
fig, axs = plt.subplots(figsize=(10,10), ncols= 3, nrows=3)
axs[0][0].plot(x, y, label='0, 0')
axs[0][1].plot(x, y, label='0, 1')
axs[0][2].plot(x, y, label='0, 2')
axs[1][0].plot(x_1, y_1, label='1, 0',marker='+')
axs[1][1].plot(x, y, label='1, 1')
axs[1][2].plot(x, y, label='1, 2')
axs[2][0].plot(x, y, label='2, 0')
axs[2][1].plot(x, y, label='2, 1')
axs[2][2].plot(x, y, label='2, 2')
plt.show()
```

```python
plt.figure(figsize=(12,5))

# scatter 산점도 만들기(x값, y값, 데이터)
plt.scatter(x='carat', y='price', data=df)
plt.show()
```

```python
# 데이터 로그변환 처리
df['log_price'] = np.log1p(df['price'])
df['log_carat'] = np.log1p(df['carat'])
```

```python
fig, axs = plt.subplots(figsize=(12,5), ncols=2, nrows=2)

# 히스토그램 그리기(칼럼명, 데이터)
axs[0][0].hist('carat', data=df)
axs[0][1].hist('price', data=df)

# 로그변환한 데이터 넣어주기(color 설정)
axs[1][0].hist('log_carat', data=df, color='g')
axs[1][1].hist('log_price', data=df, color='g')
plt.show()
```

```python
plt.figure(figsize=(12,5))

# scatter 산점도 그리기(x값, y값, 데이터)
plt.scatter(x='log_carat', y='log_price', data=df)
plt.show()
```
