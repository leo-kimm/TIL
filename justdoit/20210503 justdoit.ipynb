# 오늘의 코딩

```python
# 두 정수 사이의 합
def solution(a, b):
    answer = 0
    if a<=b:
        for i in range(a,b+1):
            answer += i
    else:
        for i in range(b,a+1):
            answer += i
    return answer
```

```python
# 문자열 내림차순으로 배열하기
def solution(s):

    answer = reversed(sorted(s))
    return ''.join(answer)

solution("Zbcdefg")
```

```python
# 문자열 내 마음대로 정렬하기 ver.1
def solution(strings, n):

    strings.sort(key=lambda strings:strings[n])
    return strings

solution(["sun", "bed", "car"],1)
```

```python
# 문자열 내 마음대로 정렬하기 ver.2
def solution(strings, n):
    strings.sort()
    strings.sort(key=lambda strings:strings[n])

    return strings

solution(["abce", "abcd", "cdx"],2)
```

```python
# 어휴 빨리 좀 만들지 함수
def get_api_result(keyword, display, start):
    
    # url 넣어주고
    url = 'https://openapi.naver.com/v1/search/image?query='+keyword+'&display='+str(display)+'&start'+str(start)
    
    # 결과값 저장하고
    result = requests.get(url, headers = api_id_pw)
    
    # 이미지 검색결과 리턴
    return result.json()

```

```python
# 함수 돌려봅시다. 피자를 100개 검색 901번째 항목 부터
get_api_result('피자',100,901)
```

```python
# 이제야 만드는 이미지 불러오기 함수
def call_and_print(keyword, total_page=10):
    # 이미지를 다운 받을 링크만 따와서 리스트에 저장
    link_list=[]
    
    # 페이지 만큼 돌려돌려
    for page in range(1,total_page+1):
        # 페이지당 100개 이미지
        display = 100
        # 페이지 시작 인덱스(1page =1, 2page =101 ...)
        start = ((page-1)*display)+1
        
        # 위에 만든 함수값 받아오기
        json_obj = get_api_result(keyword, display, start)
        
        # 링크주소만 뽑아서 리스트에 쇽
        for item in json_obj['items']:
            link_list.append(item['link'])
            
    return link_list
```

```python
# 함수 돌려~ 피자 검색해서 링크만 따오기
call_and_print('피자')
```
